## Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

## Mean imputation is the simplest imputation technique. It involves replacing missing values with the mean of the observed values for that feature. This is a quick and easy technique, but it can be problematic if the distribution of the data is not normal. For example, if the data is skewed, then the mean imputation will tend to underestimate the values of the missing data.

## Mean imputation: 
## Advantages:Quick and easy
## Disadvantages: Not robust to outliers

```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")

for i in df.columns:
    print(i, df[i].isnull().sum())
    
df = df.fillna(df.mean())

for i in df.columns:
    print(i, df[i].isnull().sum())
```
